## Set Operations

union

intersection

difference

complementary

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(set_a.union(set_b))
print(set_a | set_b | {8, 9, 10})
```

```python
set_a.union([3, 4, 5])
```

## intersect

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
print(set_a.intersection(set_b))
print(set_a & set_b)
```

## Difference

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(set_a.difference(set_b)) #A-B
print(set_b.difference(set_a)) #B-A
print(set_a - set_b) #A-B
print(set_b - set_a) #B-A      ##can only use - 
```

## Complementary

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(set_a.symmetric_difference(set_b))
```
