## Slice


slice[a:b:c]

```python
slice = "I love Bangkok!"
# "I love" -> 0,1,2,3,4,5 -> range (0, 6)
print(slice[0:6])

# Bangkok -> range (7, 14)
print(slice[7:14])

print(slice[1:8:2])
print(slice[-3:-10:-2])
print(slice[::2])
print(slice[::-1])
```
