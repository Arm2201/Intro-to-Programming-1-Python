```python

def square_number(x):
    """
    x: int - inputnumber to be squared
    Square a number then return the result.
    """ # this is called function documentation
    result = x ** 2
    return result
```

```python
# multiple inputs
import math
def calculate_gcd(num1, num2):
    gcd_number = math.gcd(num1, num2)
    return gcd_number

def give_me_6_7():
    return 6, 7, 8
    # or just use print and end with just return with nothing that follows
```

## Defult Value

```python
def f(x, y, z=7): # x, y, z are parameters
    result = x + y + z
    return result
```

```python
f(2, 4, 6) # the number inside are arguements
# value of z will be define as 7 when we remove 6 in the function call
```

```python
def f(x, y, z=6):
    result = x + y + z
    print(x, y, z)
    return result
```

```python
f(x=1, y=2, z=3)
```

```python
def add_numbers(a, b, c, d, e):
    return a + b + c + d + e
add_numbers(4, 5, 6, 8, 10)
```

```python
def product_numbers_v2(a, b, *args):
    print(args)
    result = a * b
    for num in args:
        result *= num
    return result

product_numbers_v2(1, 2, 3, 4, 5)
```

```python
def print_names(**kwargs): #kwargs is a variable so it can be change but normally we use kwargs
    print(kwargs)
    for key, value in kwargs.items():
        print(f"{key}: {value}")

        #kwargs is a dictionary that holds key value pairs like {'first_name': 'Alice', 'last_name': 'Smith'} which has an output as Alice Smith
```

```python
def my_function(a, b, c=1, d=5, *args, e=8, **kwargs):
    print(a, b, c, d, e, args, kwargs)

my_function(1, 5, 6, 8, 10, 7, f=6, g=0) #f and g are key value pairs in kwargs
```

## Literal

```python
def add_numbers(a: int, b: int, c: int) -> float:
    return (a + b) // c

add_numbers(5, 10, 3)
```

```python
def double(x: float=7) -> float:
    return x * 2

double()
```

```python
def lower_all_character(word: str="DEFAULT WORD") -> str:
    return word.lower()

lower_all_character()
```
